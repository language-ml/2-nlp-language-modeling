```python
import random 
import numpy as np
import nltk
import pandas as pd
import codecs
import tqdm
    
# Hafez and Saadi collection -- Hafez [0:8384] Saadi[9384::]
HAFEZ_end=8384
SAADI_start=9384
mesra_collection = [x.strip().split() for x in tqdm.tqdm(codecs.open('farsi/mesra.txt','rU','utf-8').readlines())]
```

<h4 style='direction:rtl;'> مثال بررسی اولیه داده غزلیات فارسی. برای مطالعه بیشتر به مقاله زیر مراجعه کنید  </h4>

<h2 style='direction:rtl;'> § نرمالایز کردن  </h2> 



```python
from __future__ import unicode_literals
from hazm import *

normalizer = Normalizer()

mesra_normalized = [[normalizer.normalize(y) for y in x] for x in tqdm.tqdm(mesra_collection)]
```

<h2 style='direction:rtl;'> § جمله‌بندی  </h2> 



```python
mesra_sentences = [sent_tokenize(' '.join(x)) for x in tqdm.tqdm(mesra_normalized)]
```

<h2 style='direction:rtl;'> § توکنایزیشن  </h2> 



```python
mesra_tokens = [[word_tokenize(sent) for sent in sents][0] for sents in tqdm.tqdm(mesra_sentences)]
```

# PMI for phrases

```python
# a more detailed code: https://github.com/asahala/Pmizer
import itertools
from collections import Counter, defaultdict
import nltk
from nltk.collocations import *
from nltk.tokenize import word_tokenize

unigrams_counts = Counter(itertools.chain(*mesra_tokens))
bigram_measures = nltk.collocations.BigramAssocMeasures()
finder = BigramCollocationFinder.from_words(list(itertools.chain(*mesra_tokens)))

for (x,y), z in finder.score_ngrams(bigram_measures.pmi)[0:500000]:
    if unigrams_counts[x]> 100 and unigrams_counts[y]>100:
        print (x,y,z)
```

# Text Representation

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(use_idf = True, norm ='l2', ngram_range=(1,1), analyzer='word')
```

```python
doc_term_mat = vectorizer.fit_transform([' '.join(doc) for doc in mesra_tokens])
words = vectorizer.get_feature_names()
```

```python
doc_term_mat.shape
```

```python
vectorizer.idf_.shape
```

```python
word_idx = vectorizer.get_feature_names()
```

```python
doc_term_mat.shape
```

# SVD embedding

```python
from scipy.sparse import csc_matrix
from scipy.sparse.linalg import svds, eigs
```

```python
# embedding creation
u, s, vt = svds(doc_term_mat.T, k=100)
svd_emb = u.dot(np.diag(s))
```

```python
import random

random_words = random.sample( words, 100)
```

```python
# get words
q_words = random_words+['دوست','یار','جنگ','مست','عشق','خاک','دشمن','وفا','فلک','قضا','بازی','کتاب','یوسف','شیرین','درس','نامه','صبا','تیر','بداندیش']
q_idx = [word_idx.index(q) for q in q_words]
```

```python
# specific embedding
q_emb = [svd_emb[ix,:] for ix in q_idx]
```

# visualize it

```python
from sklearn.manifold import TSNE
import plotly
import plotly.graph_objs as go


q_2D = TSNE(n_components = 2, random_state=0).fit_transform(q_emb)
```

```python
# Configure the layout.
layout = go.Layout(
    margin = {'l': 0, 'r': 0, 'b': 0, 't': 0},
    showlegend=True,
    hoverlabel=dict(
        bgcolor="white", 
        font_size=20, 
        font_family="Courier New"),
    legend=dict(
    x=1,
    y=0.5,
    font=dict(
        family="Courier New",
        size=25,
        color="black"
    )),
    font = dict(
        family = " Courier New ",
        size = 15),
    autosize = False,
    width = 1000,
    height = 1000
    )

trace = go.Scatter(
                    x = q_2D[:,0], 
                    y = q_2D[:,1],  
                    text = q_words,
                    textposition = "top center",
                    textfont_size = 20,
                    mode = 'markers+text',
                    marker = {
                        'size': 15,
                        'opacity': 0.8,
                        'color': 2
                    }
       
                )
plot_figure = go.Figure(data = [trace], layout = layout)
```
